```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
#drop unrelated/blank columns
df.drop(['Status','unnamed1'], axis = 1, inplace = True) #axis=1: This specifies that you want to drop columns, not rows. The value 1 refers to columns; if you wanted to drop rows, you would use axis=0 /// inplace = True is used to save in existing 
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace = True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int') # data type of amount is change from float to int
```

<h1>Exploratory Data Analysis</h1>

<h3> Gender </h3>

```python
# plotting a bar chart for Gender and it's count
ax = sns.countplot(x = 'Gender', data = df)
sns.set(rc = {'figure.figsize':(14,5)})
# this loop is used to show the count values on the bar
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False) # ascending = if false then it sort values in descending order
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

<h5>From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men</h5>

<h3> Age Group </h3>

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender') # hue is used to catogarize data in same plot
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_gen = df.groupby(['Age Group', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_gen, hue = 'Gender')
```

<h5>From above graphs we can see that most of the buyers are of age group between 26-35 yrs female</h5>

 <h3> State </h3>

```python
# total number of orders from top 10 states
sales_gen = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(19,5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_gen)
```

```python
# total amount/sales 
sales_gen = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize': (20,5)})
sns.barplot(x = 'State', y = 'Amount', data = sales_gen)
```

<h5>From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively </h5>

<h3> Marital Status </h3>

```python
ax = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc = {'figure.figsize' : (7,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize': (6,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_gen, hue ='Gender')
```

<h5>From above graphs we can see that most of the buyers are married (women) and they have high purchasing power</h5>

<h3> Occupation </h3>

```python
ax = sns.countplot(x = 'Occupation', data = df)
sns.set(rc= {'figure.figsize': (28.5,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Occupation', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize' : (20,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_gen, hue = 'Gender')
```

```python
sales_gen = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize' : (20,5)})
sns.barplot(x = 'Occupation', y = 'Amount', data = sales_gen)
```

<h5>From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector</h5>

<h3> Product Category </h3>

```python
ax = sns.countplot(x = 'Product_Category', data = df)
sns.set(rc = {'figure.figsize' : (20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(x = 'Product_Category', y = 'Amount', data = sales_gen)
```

<h5>From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category</h5>


<h3> Product ID </h3>

```python
sales_gen = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize' : (20,5)})
sns.barplot(x = 'Product_ID', y = 'Orders', data = sales_gen)
```
